```python
import random

def f(x):
  return x ** 2

def generate_samples(n = 10):
  samples = [random.uniform(0, 1) for _ in range(n)]
  return samples

samples = generate_samples(10)

f_values = [f(x)  for x in samples]

i_estimate = sum(f_values)/len(f_values)
print(samples)
print(f_values)
print(i_estimate)
```
