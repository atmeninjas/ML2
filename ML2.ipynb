```python
!pip install orange3
import Orange

data = Orange.data.Table("iris")


cn2 = Orange.classification.rules.CN2Learner()
cn2_rules = cn2(data)

foil = Orange.classification.rules.CN2SDUnorderedLearner()
foil_rules = foil(data)

print("=== CN2 RULES ===")
for r in cn2_rules.rule_list:
    print(r)

print("\n=== FOIL-LIKE RULES ===")
for r in foil_rules.rule_list:
    print(r)
```
