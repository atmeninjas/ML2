```python
def find_s(data):
    S = ['0'] * (len(data[0]) - 1)
    for x in data:
        a, y = x[:-1], x[-1].lower()
        if y == 'yes':
            if S == ['0'] * len(a):
                S = a.copy()
            else:
                S = ['?' if S[i] != a[i] else S[i] for i in range(len(a))]
    return S

def more_general(g, s):
    return all(g[i] == '?' or g[i] == s[i] for i in range(len(g)))

def candidate_elimination(data):
    n = len(data[0]) - 1
    D = [[x[i] for x in data] for i in range(n)]
    S, G = ['0'] * n, [['?'] * n]
    for x in data:
        a, y = x[:-1], x[-1].lower()
        if y == 'yes':
            G = [g for g in G if all(g[i] == '?' or g[i] == a[i] for i in range(n))]
            if S == ['0'] * n:
                S = a.copy()
            else:
                S = ['?' if S[i] != a[i] else S[i] for i in range(n)]
            G = [g for g in G if more_general(g, S)]
        else:
            newG = []
            for g in G:
                if all(g[i] == '?' or g[i] == a[i] for i in range(n)):
                    for i in range(n):
                        if g[i] == '?':
                            for val in set(D[i]) - {a[i]}:
                                h = g.copy()
                                h[i] = val
                                newG.append(h)
                else:
                    newG.append(g)
            G = [h for h in newG if not any(more_general(h2, h) and h != h2 for h2 in newG)]
    return S, [g for g in G if more_general(g, S)]

data = [
    ['Sunny', 'Warm', 'Normal', 'Yes'],
    ['Sunny', 'Warm', 'High',   'Yes'],
    ['Rainy', 'Cold', 'High',   'No'],
    ['Sunny', 'Warm', 'High',   'Yes']
]

S, G = candidate_elimination(data)
print("Find-S Hypothesis:", find_s(data))
print("Most Specific (S):", S)
print("Most General (G):", G)
```
